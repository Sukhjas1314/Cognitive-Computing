```python
import numpy as np
gfg = np.matrix('[4, 1, 9; 12, 3, 1; 4 , 5, 6]')
sum = np.sum(gfg)
print(f'Sum of all elements : {sum}')
```

##### (¡¡)

```python
sum_rows = np.sum(gfg,axis = 1)
print(f'Sum of all elements row-wise : \n{sum_rows}')
```

##### (¡¡¡)

```python
sum_column = np.sum(gfg,axis = 0)
print(f'Sum of all elements column-wise : {sum_column}')
```

### Q2.

##### (a)

```python
array = np.array([10,52,62,16,16,54,453])

# (¡)
sorted_array = np.sort(array)
print(f'Sorted array : {sorted_array}\n')

# (¡¡)
indices_sorted_arr = np.argsort(array)
print(f'Indices of sorted array : {indices_sorted_arr}\n')

# (¡¡¡)
smallest_4 = np.sort(array)[:4]
print(f'4 smallest elements from the array : {smallest_4}\n')

# (¡v)
largest_5 = np.sort(array)[-5:]
print(f'5 largest elements from the array : {largest_5}\n')
```

##### (b)

```python
array1 = np.array([1.0,1.2,2.2,2.0,3.0,2.0])

# (¡)
integers = array1[array1 == array1.astype(int)]
print(f'Integers elements only : {integers}\n')

# (¡¡)
float_elements = array1[array1 != array1.astype(int)]
print(f'Float elements only : {float_elements}\n')
```

### Q3.

```python
# (a) 
first_initial, last_initial = 'S', 'T'
X = ord(first_initial) + ord(last_initial)
print("Sum of initials :",X)

sales = np.array([X, X+50, X+100, X+150, X+200])
print("\nSales Data :",sales)

# (b)
tax_rate = ((X % 5) + 5) / 100
taxed_sales = sales * (1 + tax_rate)
print("\nTaxed Sales :",taxed_sales)

# (c) 
discounted_sales = np.where(sales < X+100,sales * 0.95,sales * 0.9)
print("\nDiscounted Sales :",discounted_sales)

# (d) 
weekly_sales = np.zeros((3, 5))
weekly_sales[0] = sales  
weekly_sales[1] = sales * 1.02 
weekly_sales[2] = sales * 1.04
print("\nWeekly Sales Data :\n",weekly_sales)
```

### Q4.

```python
import matplotlib.pyplot as plt

x = np.linspace(-10, 10, 100)
print("x :",x)
print()

y1 = x**2
print("y1 :",y1)
print()

y2 = np.sin(x)
print("y1 :",y1)
print()

y3 = np.exp(x)
print("y1 :",y1)
print()

y4 = np.log(np.abs(x) + 1)
print("y1 :",y1)
print()
```

```python
plt.plot(x, y1, label="y = x^2")
plt.xlabel("x values")
plt.ylabel("y values")
plt.title("x^2 Graph")
plt.grid(True)
plt.show()
```

```python
plt.plot(x, y2, label="y = sin(x)")
plt.xlabel("x values")
plt.ylabel("y values")
plt.title("sin(x) Graph")
plt.grid(True)
plt.show()
```

```python
plt.plot(x, y3, label="y = e^x")
plt.xlabel("x values")
plt.ylabel("y values")
plt.title("e^x Graph")
plt.grid(True)
plt.show()
```

```python
plt.plot(x, y4, label="y = log(|x|+1)")
plt.ylabel("y values")
plt.title("log(|x| + 1) Graph")
plt.grid(True)
plt.show()
```
