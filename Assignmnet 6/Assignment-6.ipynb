```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### Q1.

```python
M = float(input("Enter a value : "))
x = np.linspace(-10,10,100)

y1 = M * x**2
y2 = M * np.sin(x)
```

```python
plt.figure(figsize=(8,6))
plt.plot(x,y1,label = 'M*x^2',color = 'blue',linestyle = '-')
plt.plot(x,y2,label = 'M*sin(x)',color = 'red',linestyle = '--')

plt.legend()
plt.grid(True)
plt.title('Plot of y = M*x^2 and y = M*sin(x)')
plt.xlabel('X values')
plt.ylabel('Y values')
plt.show()
```

### Q2.

```python
data = {
    'Subjects' : ['Maths','Physics','Chemistry','English','IT'],
    'Scores' : [95,96,100,97,99]
}
df = pd.DataFrame(data)

plt.figure(figsize = (8,6))
colors = sns.color_palette('husl',len(df))
sns.barplot(x='Subjects',y='Scores',data=df,hue='Subjects',palette=colors,legend=False)

for i,score in enumerate(df['Scores']):
    plt.text(i,score+1,str(score),ha='center',fontsize=12)

plt.grid(axis='y',linestyle='--',alpha=0.7)
plt.title('Score Sheet of different Subjects')
plt.xlabel('Subjects')
plt.ylabel('Scores')
plt.show()
```

### Q3.

```python
Rno = 102317112
np.random.seed(Rno)

data = np.random.randn(50)

fig,axes = plt.subplots(2,2,figsize=(10,10))

axes[0,0].plot(np.cumsum(data),color='blue',linestyle='-',marker='o',markersize=4)
axes[0,0].set_title('Cumulative Sum Plot')
axes[0,0].set_xlabel('Index')
axes[0,0].set_ylabel('Cumulative Sum')
axes[0,0].grid(True)

axes[0,1].scatter(range(50),data,color='red',alpha=0.7)
axes[0,1].set_title('Scatter Plot with Random noise')
axes[0,1].set_xlabel('Index')
axes[0,1].set_ylabel('Value')
axes[0,1].grid(True)

axes[1,0].axis('off')
axes[1,1].axis('off')

plt.show()
```

### Q4.

```python
dataset = pd.read_csv('company_sales_data.csv')

plt.figure(figsize=(10,8))
sns.lineplot(data=dataset,x='month_number',y='total_profit',marker='o',color='blue')
plt.title('Total Profit per Month')
plt.xlabel('Month Number')
plt.ylabel('Total Profit')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(12,8))
product_cols = dataset.columns[1:7]
for product in product_cols:
    sns.lineplot(data=dataset,x='month_number',y=product,marker='o',label=product)

plt.title('Product sales per Month')
plt.xlabel('Month number')
plt.ylabel('Sales units')
plt.legend(title='Products')
plt.grid(True)
plt.show()
```

```python
data_sum = dataset.sum()
plt.figure(figsize=(14,7))
sns.barplot(x=data_sum.index,y=data_sum.values,palette='viridis')
plt.title('Total Sum of each attribute')
plt.xlabel('Attributes')
plt.ylabel('Total Sum')
plt.xticks(rotation=30)
plt.grid(axis='y',linestyle='--',alpha=0.7)
plt.show()
```
